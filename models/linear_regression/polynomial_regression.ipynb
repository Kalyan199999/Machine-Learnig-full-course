```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
np.random.seed(19)
```

```python
X = 6 * np.random.rand(100,1) - 3

y = 0.5 * X**2 + 1.5*2 + 2 + np.random.randn(100,1)
```

```python
plt.scatter( X,y , color='g' )

plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split( X,y , test_size=0.25 , random_state=42 )
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
```

```python
linear_regression.fit( X_train , y_train )
```

```python
from sklearn.metrics import r2_score , mean_squared_error , mean_absolute_error
```

```python
y_test_pred =  linear_regression.predict( X_test )
```

```python
r_2_scores = r2_score( y_test , y_test_pred )

mse = mean_squared_error( y_test , y_test_pred )

mae = mean_absolute_error( y_test , y_test_pred )

print( f'Accuracy of the model:{r_2_scores}' )
print( f'mean_squared_error of the model:{mse}' )
print( f'mean_absolute_error of the model:{mae}' )
```

```python
plt.scatter( X_test ,y_test , color='g' )

plt.plot(X_test, y_test_pred, 'b' )

plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```

### Ploynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures( degree=2 , include_bias=True  )
```

```python
X_train_poly = poly.fit_transform( X_train )

X_test_poly = poly.transform( X_test )
```

```python
# X_test_poly
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit( X_train_poly , y_train )
```

```python
y_test_pred_poly =  linear_regression.predict( X_test_poly )
```

```python
r_2_scores = r2_score( y_test , y_test_pred_poly )

mse = mean_squared_error( y_test , y_test_pred_poly )

mae = mean_absolute_error( y_test , y_test_pred_poly )

print( f'Accuracy of the model:{r_2_scores}' )
print( f'mean_squared_error of the model:{mse}' )
print( f'mean_absolute_error of the model:{mae}' )
```

```python
plt.scatter( X_test ,y_test , color='g' )

plt.scatter(X_test, y_test_pred_poly, color='b' )

plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```
