```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
cat_cols = ['sex', 'smoker', 'day', 'time']

num_cols = ['total_bill', 'tip','size']
```

```python
for col in cat_cols:
    print(df[col].value_counts())
```

```python
X = df[ [ 'tip', 'sex', 'smoker', 'day', 'time', 'size'] ]

y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=43,test_size=0.25)
```

### Encoding the Categorical columns

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform( X_train['sex'] )

X_train['smoker'] = le2.fit_transform( X_train['smoker'] )

X_train['time'] = le3.fit_transform( X_train['time'] )
```

```python
X_test['sex'] = le1.transform( X_test['sex'] )

X_test['smoker'] = le2.transform( X_test['smoker'] )

X_test['time'] = le3.transform( X_test['time'] )
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(
    transformers=[ ( 'onehot',OneHotEncoder(drop='first') ,[3] ) ],
    remainder='passthrough'
)
```

```python
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

### import the model 

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
svr.fit( X_train , y_train )
```

```python
y_pred = svr.predict( X_test )
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print( f'The mean squared error is {mean_squared_error( y_test , y_pred )}' )

print( f'The r2 score is {r2_score( y_test , y_pred )}' )
```
