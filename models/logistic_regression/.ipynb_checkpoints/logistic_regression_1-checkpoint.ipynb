```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Make own dataset

```python
from sklearn.datasets import make_classification
```

```python
X , y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2, # binary classification
    random_state=42
)
```

```python
X
```

```python
y
```

### train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split( X, y, test_size=0.25 , random_state=42 )
```

### Model selection

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit( X_train , y_train )
```

```python
y_train_pred = model.predict( X_train )

y_test_pred = model.predict( X_test)
```

```python
# probability of the both class and which ever is large that is predicted value
model.predict_proba( X_test )
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report
```

```python
print( "Training predictions" )

print(f'Accuracy of the traing data: { accuracy_score( y_train , y_train_pred ) }')

print(f'Confusion matrix of the traing data: { confusion_matrix( y_train , y_train_pred ) }')
```

```python
("Classfication Report of Training")

print(classification_report( y_train , y_train_pred ))
```

```python
print( "Testing predictions" )

print(f'Accuracy of the traing data: { accuracy_score( y_test , y_test_pred ) }')

print(f'Confusion matrix of the traing data: { confusion_matrix( y_test , y_test_pred ) }')
```

```python
("Classfication Report of test")

print(classification_report(y_test , y_test_pred))
```

# Hyperparamter Tunning and cross validation

```python
penalty = ['l1' , 'l2' , 'elasticnet']
c_values = [ 100 , 10, 1, 0.1 , 0.01]

solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict( penalty=penalty , C=c_values , solver=solver )
```

```python
model = LogisticRegression(  )
```

### Grid searchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold()
```

```python
grid = GridSearchCV( estimator=model , param_grid=params , scoring='accuracy' ,cv=cv , n_jobs=-1 )
```

```python
grid
```

```python
grid.fit( X_train , y_train )
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict( X_test )
```

```python
print( "Testing predictions" )

print(f'Accuracy of the traing data: { accuracy_score( y_test , y_pred ) }')

print(f'Confusion matrix of the traing data: { confusion_matrix( y_test , y_pred ) }')

("Classfication Report of test")

print(classification_report(y_test , y_pred))
```
