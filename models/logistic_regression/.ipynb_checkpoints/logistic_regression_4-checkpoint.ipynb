```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X , y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2, # binary classification
    random_state=42
)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split( X, y, test_size=0.25 , random_state=42 )
```

```python
y_test
```

### dummy model prediction

```python
dummy_model_prob = [ 0 for _ in range(len(y_test)) ]

dummy_model_prob
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
model = LogisticRegression()
```

```python
model.fit( X_train , y_train )
```

```python
model_prob = model.predict_proba( X_test )

model_prob
```

```python
model_prob = model_prob[ : ,1 ]
model_prob
```

```python
y_test_pred = model.predict( X_test)
```

```python
dummy_model_auc = roc_auc_score( y_test , dummy_model_prob )

model_auc = roc_auc_score( y_test , model_prob )
```

```python
print(f'Dummy model accuracy is:{dummy_model_auc}')

print(f'Logistic Regression model accuracy is:{model_auc}')
```

```python
dummy_fpr , dummy_tpr ,_ = roc_curve( y_test,dummy_model_prob )

model_fpr , model_tpr ,_ = roc_curve( y_test , model_prob )
```

```python
dummy_fpr , dummy_tpr
```

```python
model_fpr , model_tpr
```

```python

plt.title('Roc curve')

plt.plot( dummy_fpr , dummy_tpr , linestyle='-' , label='Dummy Model' )

plt.plot(model_fpr , model_tpr, marker='.' , label='Logistic Regression Model' )

plt.xlabel('False Postive rate')
plt.ylabel('True Postive rate')

plt.legend()

plt.show()
```

### from the above curve the area under Logistic regression is more than the Dummy Model

```python
dummy_fpr , dummy_tpr ,_ = roc_curve( y_test,dummy_model_prob )

model_fpr , model_tpr ,thresholds = roc_curve( y_test , model_prob )
```
