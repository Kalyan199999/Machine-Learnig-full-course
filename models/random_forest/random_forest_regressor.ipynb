### get categorical and numerical columns and descrete and continuous features

### use label encoding for model and one hot encoding for seller_type fuel_type transmission_type

### Encoding the categorical features and standardingbthe numerical columns

```python
from sklearn.preprocessing import OneHotEncoder , LabelEncoder ,StandardScaler
```

```python
ohe = OneHotEncoder( drop='first' , dtype=int )
le = LabelEncoder()
scaler = StandardScaler()
```

```python
df.head()
```

```python
df['model'] = le.fit_transform( df['model']  )
```

```python
df.head()
```

```python
num_features = num_cols[:-1]
num_features
```

```python
cat_features = cat_cols[1:]
cat_features
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("OneHotEncoder",ohe,cat_features),
        ("StandardScaler",scaler,num_features)
    ],
    remainder='passthrough'
)
```

### seperate the independent and dependent columns

```python
X = df.iloc[:,:-1]
y = df.iloc[ : ,-1 ]
```

```python
X
```

```python
X_trasformed = preprocessor.fit_transform(X)
pd.DataFrame(X_trasformed)
```

### train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split( X_trasformed,y , test_size=0.25, random_state=43 )
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit( X_train , y_train )
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
def evaluation_metrics( y_true , y_prediction ):
    r2 = r2_score( y_true , y_prediction )
    mse = mean_squared_error( y_true , y_prediction )
    mae = mean_absolute_error( y_true , y_prediction )
    
    print(f'R2 score: {r2}')
    print(f'Mean squared error: {mse}')
    print(f'Mean Absolute error:{mae}')
    
    return r2,mse,mae
```

```python
y_pred = rfr.predict( X_test )
```

```python
evaluation_metrics( y_test,y_pred )
```
