```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop( columns='prodtaken' , axis=1  )

y = df['prodtaken']
```

```python
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size=0.25 , random_state=42 )
```

```python
X_train.shape , X_test.shape
```

### transform the data

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler

from sklearn.compose import ColumnTransformer
```

```python
scaler = StandardScaler()

encoder = OneHotEncoder( drop='first' )
```

```python
cat_cols = X.select_dtypes(include='object').columns

num_cols = X.select_dtypes(exclude='object').columns
```

```python
preprocessor = ColumnTransformer(
    [
        ( 'OneHotEncoder' , encoder, cat_cols  ),
        ( 'StandardScaler' , scaler, num_cols )
    ]
)
```

```python
X_train_processed = preprocessor.fit_transform( X_train )
```

```python
X_test_processed = preprocessor.transform( X_test )
```

### Model

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit( X_train_processed , y_train ) 
```

```python
y_pred = rfc.predict( X_test_processed )
```

```python

print('Confusion matrix')
print(confusion_matrix(y_test , y_pred ))

print(f'Accuracy score is:{accuracy_score(y_test , y_pred)}')

print('Classification Report')
print(classification_report(y_test , y_pred ))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# test on different models

models = {
    "Logistic regression":LogisticRegression(),
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
}
```

```python
for i in range( len( list(models) ) ):
    model = list(models.values())[i]
    model.fit( X_train_processed , y_train )
    
    y_train_pred = model.predict(  X_train_processed )
    
    y_test_pred = model.predict(  X_test_processed )

    

    print(f'Prediction for the model {list(models)[i]}')
    
    # Training Prediction
    print('Training Predictions')
    print('Confusion matrix')
    print(confusion_matrix(y_train , y_train_pred ))
    print(f'Accuracy score is:{accuracy_score(y_train , y_train_pred)}')
    print('Classification Report')
    print( classification_report(y_train , y_train_pred )  )

    print('----------------------------------------------------------------------------------------------------------------------------------------------------')
          
    # Testing Prediction
    print('Training Predictions')
    print('Confusion matrix')
    print(confusion_matrix(y_test , y_test_pred ))
    print(f'Accuracy score is:{accuracy_score(y_test , y_test_pred)}')
    print('Classification Report')
    print(classification_report(y_test , y_test_pred ))
          
    print("===================================================================================================================================================================")
    
    
```
