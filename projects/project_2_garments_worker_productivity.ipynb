```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop(columns='productivity')
y = df['productivity']
```

```python
X_train,X_test,y_train,y_test = train_test_split( X,y, test_size=0.25 , random_state=42 )
```

```python
cat_cols = [ col for col in X_train.columns  if df[col].dtype == 'O' ]
num_cols = [ col for col in X_train.columns  if df[col].dtype != 'O' ]
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder(dtype='int' , drop='first')
scale = StandardScaler()
```

```python
ct = ColumnTransformer(transformers=
    [
      ('OneHotEncoder' , ohe , cat_cols) , 
      ('StandardScaler' , scale , num_cols)
   ],
    remainder='passthrough',
)
```

```python
X_train
```

```python
X_train_ct = ct.fit_transform( X_train )
```

```python
X_test_ct = ct.transform( X_test )
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
rfc = RandomForestClassifier(  )

dtc = DecisionTreeClassifier()
```

```python
rfc.fit( X_train_ct , y_train )
```

```python
dtc.fit( X_train_ct , y_train )
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

## for Random forest classification

```python
rfc_train_pred = rfc.predict( X_train_ct )
rfc_test_pred = rfc.predict( X_test_ct )
```

```python
print( 'For training data ')

print( f'Accuracy of traing data:{accuracy_score( y_train , rfc_train_pred )}')

print( f'Confucsion Matrix: {confusion_matrix( y_train , rfc_train_pred )}')

print( f'Classification report: {classification_report( y_train , rfc_train_pred )}')
```

```python
print( 'For testing data ')

print( f'Accuracy of testing data:{accuracy_score( y_test , rfc_test_pred )}')

print( f'Confucsion Matrix: {confusion_matrix( y_test , rfc_test_pred )}')

print( f'Classification report: {classification_report( y_test , rfc_test_pred )}')
```
