```python
image = cv2.imread( os.path.join( root_dir , image_list[2] ) )
```

```python
def displayImage( src ):
    plt.title('pichuka')
    plt.imshow(src)
    plt.axis(False)
    plt.show()

displayImage(image)
```

```python
gray_image = cv2.cvtColor( image , cv2.COLOR_BGR2GRAY )

displayImage(gray_image)
```

```python
_,image_thresold = cv2.threshold( gray_image , 120 , 255 ,cv2.THRESH_BINARY)

displayImage(image_thresold)
```

```python
displayImage(cv2.cvtColor(image_thresold,cv2.COLOR_GRAY2RGB ))
```

```python
image_edge = cv2.Canny( gray_image , 100,255 )
displayImage( image_edge )
```

```python
displayImage(cv2.cvtColor(image_edge,cv2.COLOR_GRAY2RGB ))
```

```python
image_dilate = cv2.dilate( image_edge , np.ones( (5,5) , dtype=np.int8 ) )

displayImage(cv2.cvtColor(image_dilate,cv2.COLOR_GRAY2RGB ))
# displayImage(image_dilate)
```
