```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
path = '../All_data_sets/dataset'
```

```python

img_paths = os.listdir( path )

img_paths = img_paths[1:-1]

img_paths
```

```python
all_images = []

for img in img_paths:
    all_images.append( cv2.imread( os.path.join( path , img ) ) )
```

```python
def display( image , title='Image' ):
    plt.title(title)
    plt.imshow( image )
    plt.axis(False)
    plt.show()
```

```python
image_5 = cv2.cvtColor( all_images[5] , cv2.COLOR_BGR2RGB )

def contoursImage( image ):
    image_gray = cv2.cvtColor( image , cv2.COLOR_BGR2GRAY )
    
    _ , thresold = cv2.threshold( image_gray , 100,255,cv2.THRESH_BINARY_INV )   
    contours , hierarchy = cv2.findContours( thresold , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE )
    
    for cnt in contours:
        if( cv2.contourArea( cnt ) > 0 ):
            cv2.drawContours( image , cnt , -1 , ( 0,255,0 ) , 1 )

    display(image)

contoursImage(image_5)
```
