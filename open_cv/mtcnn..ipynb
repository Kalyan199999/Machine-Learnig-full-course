```python
from mtcnn import MTCNN
```

```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
root_dir = '../All_data_sets/dataset'
```

```python
image_list = os.listdir(root_dir)[1:]

image_list
```

```python
def displayImage( src ):
    plt.title('pichuka')
    plt.imshow(src)
    plt.axis(False)
    plt.show()
```

```python
# Load image
mb1 = cv2.imread( os.path.join( root_dir,image_list[6]) ) 
mb1_rgb = cv2.cvtColor(mb1, cv2.COLOR_BGR2RGB)

displayImage(mb1_rgb)
```

```python
# Load image
mb2 = cv2.imread( os.path.join( root_dir,image_list[7]) ) 
mb2_rgb = cv2.cvtColor(mb2, cv2.COLOR_BGR2RGB)

displayImage(mb2_rgb)
```

```python
# Initialize detector
detector = MTCNN(device="GPU:1")
```

```python
# Detect faces
faces_1 = detector.detect_faces(mb1_rgb)

faces_1
```

```python
# Draw detections
resize_mb_1 = []
for face in faces_1:
    x, y, w, h = face['box']
    resize_mb_1 = mb1_rgb[ y:y+h , x:x+w ]
    cv2.rectangle(mb1_rgb, (x, y), (x+w, y+h), (0,255,0), 2)
    for key, point in face['keypoints'].items():
        cv2.circle(mb1_rgb, point, 2, (255,0,0), -1)
```

```python
displayImage( mb1_rgb )
```

```python
displayImage( resize_mb_1 )
```
