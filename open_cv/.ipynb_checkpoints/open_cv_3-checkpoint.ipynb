# Thresholding & Masking.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = "../All_data_sets/dataset"
```

```python
os.listdir(path)
```

```python
def display(img):
    
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
image = cv2.imread( path+'/mb_2.jpeg' )
```

## This step is important because it helps in segmenting images, extracting regions of interest, and preparing for contour/object detection.

## Simple Thresholding

## Turns the image into black & white based on a threshold value.

```python
# Apply simple threshold
_, th1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
_, th2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)


titles = ['Original', 'Binary', 'Binary Inv']
images = [image, th1, th2]
```

```python

for i in range(3):
    plt.subplot(1,3,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i]), plt.axis("off")
plt.show()
```

## Adaptive Thresholding

## Useful when lighting is uneven in the image.

```python
# pip install opencv-python-headless
```

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Adaptive threshold
th3 = cv2.adaptiveThreshold(gray, 255,
                            cv2.ADAPTIVE_THRESH_MEAN_C,
                            cv2.THRESH_BINARY, 11, 2)

th4 = cv2.adaptiveThreshold(gray, 255,
                            cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv2.THRESH_BINARY, 11, 2)

# Show results
titles = ['Original', 'Adaptive Mean', 'Adaptive Gaussian']
images = [gray, th3, th4]
```

```python

for i in range(3):
    plt.subplot(1,3,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i]), plt.axis("off")
plt.show()
```

## Masking with Bitwise Operations

## We can extract parts of the image using masks.

```python
# Create a mask (circle in center)
mask = np.zeros(gray.shape, dtype="uint8")

cv2.circle(mask, (gray.shape[1]//2, gray.shape[0]//2), 100, 255, -1)

# Apply mask
masked = cv2.bitwise_and(image, image, mask=mask)

masked_gray = cv2.bitwise_and(gray, gray, mask=mask)
```

```python
plt.subplot(1,3,1)
plt.imshow(mask, cmap='gray')
plt.title("Mask")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))  # Convert BGR  RGB
plt.title("Masked Color")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(masked_gray, cmap='gray')
plt.title("Masked Gray")
plt.axis('off')

plt.show()
```
