# Lammatization

```python
import pandas as pd
import numpy as np
```

## Lemmatization reduces words to their dictionary form (lemma) using grammar and meaning.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
words = ["running", "played", "better", "studies", "mice" , "history" , "convencing" , "brother" , "eating" , "ate" ]
```

```python
df = pd.DataFrame(data={
    "word":words
})
```

```python
df
```

### Parts of speech is verb(v)

```python
df['lemma_verb'] = df['word'].apply( lambda x : lemma.lemmatize(x , pos='v') )
```

```python
df
```

### Parts of speech is Noun(n)

```python
df['lemma_noun'] = df['word'].apply( lambda x : lemma.lemmatize(x , pos='n') )
```

```python
df
```

### Parts of speech is Adjective(a)

```python
df['lemma_adj'] = df['word'].apply( lambda x : lemma.lemmatize(x , pos='a') )

df
```

# Comping with setemming

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
df['stemming'] = df['word'].apply( lambda x : ps.stem(x ) )

df
```
