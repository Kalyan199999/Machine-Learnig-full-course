```python
path = '../../All_data_sets/pandas_dataset/taitanic.csv'
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(path)

df.head()
```

### create the new columns

```python
df['First_name'] = df['Name'].str.split(',').str[0]

df.head()
```

### rename the columns

```python
df.rename(columns={ 'Sex':'Gender' } , inplace=True)

df.head()
```

```python
df.rename( columns=str.lower , inplace=True)

df.head()
```

# How to calculate summary statistics

```python
df.describe()
```

### What is the average age of the Titanic passengers?

```python
df['age'].mean()
```

### What is the median age and ticket fare price of the Titanic passengers?

```python
df[ ['age','fare'] ].median()
```

### specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:

```python
df.agg(
    {
        "age": ["min", "max", "median", "skew"],
        "fare": ["min", "max", "median", "mean"],
    }
)
```

# Aggregating statistics grouped by category

### What is the average age for male versus female Titanic passengers?

```python
df[["gender", "age"]].groupby("gender").mean()
```

```python
df.groupby('gender')['age'].mean()
```

```python
df[["gender", "age",'pclass']].groupby("gender").mean()
```

```python
df.groupby('gender')[['age','pclass']].mean()
```

### What is the mean ticket fare price for each of the sex and cabin class combinations?

```python
df.groupby(['gender','pclass'])[['fare']].mean()
```

### What is the number of passengers in each of the cabin classes?

```python
df['pclass'].value_counts()
```

```python
df.groupby('pclass')['pclass'].count()
```

### sort the data by age columns

```python
df.sort_values(by='age').head()
```

```python
df.sort_values(by='age',ascending=False).head()
```

### sort the Titanic data according to the cabin class and age in descending order.


```python
df.sort_values( by=['pclass', 'age'], ascending=[True,False] ).head()
```
