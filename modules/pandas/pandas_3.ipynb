### How to select a subset of a DataFrame

```python
# for single column

df['Name']
```

### for multiple columns

```python
# To select multiple columns, use a list of column names within the selection brackets [].

df[ [ 'Name'	,'Sex'	,'Age' ]  ]
```

#### How to filter specific rows from a DataFrame

```python
df['Age']>35 
```

```python
df[ df['Age']>35 ]
```

```python
df[ df['Pclass'].isin([1,3]) ]
```

### When combining multiple conditional statements, each condition must be surrounded by parentheses (). Moreover, you can not use or/and but need to use the or operator | and the and operator &.

```python
df[ (df['Pclass'] == 1) | (df['Pclass']==3)  ]
```

### How to select the not null values

```python
df[df['Survived'].notna()]
```

### How to select specific rows and columns from a DataFrame

### The loc/iloc operators are required in front of the selection brackets []. When using loc/iloc, the part before the comma is the rows you want, and the part after the comma is the columns you want to select.

### Select specific rows and/or columns using loc when using the row and column names.

### Select specific rows and/or columns using iloc when using the positions in the table.

```python
df.loc[ df['Age']>35 , 'Name' ]
```

```python
df.loc[ df['Age']>35 , ['Name','Age'] ]
```

```python
df.loc[ 0:10,['Name','Age'] ]
```

```python
df.iloc[ 100:200 , 3:6 ]
```

```python
df.iloc[ : , 3: ]
```
