### Nomial Encoder or one-hot encoder

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(
    {
        'color':['red','blue','green','red','red','blue','blue','green','red']
    }
)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

### assigns the values based on the alphabetical order

```python
ohe_spare_matrix = ohe.fit_transform(df[['color']])

print(ohe_spare_matrix)
```

```python
ohe_spare_matrix.toarray()
```

```python
ohe.transform( [['red']] ).toarray()
```

```python
ohe.transform( [['blue']] ).toarray()
```

```python
ohe.get_feature_names_out()
```

```python
colored_df = pd.DataFrame(
    data=ohe_spare_matrix.toarray(),
    columns=ohe.get_feature_names_out()
)
```

```python
colored_df.head()
```

```python
final_df = pd.concat(
    [df,colored_df],
    axis=1
)
```

```python
final_df
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
ohe = OneHotEncoder()
```

```python
sex = ohe.fit_transform( df[['sex']] ).toarray()
```

```python
df_sex = pd.DataFrame(
    sex ,
    columns=ohe.get_feature_names_out()
)
```

```python
df = pd.concat(
    [df , df_sex],
    axis=1
)
```

```python
df.drop(columns='sex',inplace=True)
```

```python
df.head()
```

```python
smoker = ohe.fit_transform( df[['smoker']] ).toarray()
```

```python
df_smoker = pd.DataFrame(
    smoker,
    columns=ohe.get_feature_names_out()
)
```

```python
df_smoker
```

```python
df = pd.concat(
    [df , df_smoker],
    axis=1
)
```

```python
df.drop(columns='smoker' , inplace=True)
```

```python
df
```
