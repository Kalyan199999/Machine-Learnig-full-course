### imputation techniques(handling missing values)

### 1-Mean value impute

```python
sns.histplot(df['age'] , kde=True)
plt.show()
```

```python
df['age'].mean()
```

```python
df['age_mean'] = df['age'].fillna( df['age'].mean() )
```

```python
df[['age' , 'age_mean']]
```

```python
sns.histplot(df['age_mean'] , kde=True)
plt.show()
```

```python
df[ (df['age_mean'] == df['age'].mean()) ][['age' , 'age_mean']]
```

### 2-Meadian imputation

```python
df['age_median'] = df['age'].fillna( df['age'].median() )
```

```python
df[['age','age_median']]
```

```python
df[ (df['age_median'] == df['age'].median()) ][['age' , 'age_median']]
```

### 3-Mode value impute

```python
sns.histplot(df['embarked'] , kde=True)
plt.show()
```

```python
mode_val = df[ df['embarked'].notna() ]['embarked'].mode()[0]
```

```python
mode_val
```

```python
df['embarked_mode'] = df['embarked'].fillna( mode_val )
```

```python
df[ (df['embarked'].isnull()) ][['embarked' , 'embarked_mode']]
```
