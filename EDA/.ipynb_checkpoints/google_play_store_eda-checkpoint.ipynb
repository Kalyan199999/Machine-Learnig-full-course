# Data cleaning

```python
df.rename(columns=str.lower , inplace=True)
```

### column : rating

```python
df['rating'].unique()
```

### handling the null values with the mean

```python
df[ df['rating'].isnull() ]
```

```python
df['rating'].fillna( df['rating'].mean() , inplace=True )
```

```python
df[ df['rating'].isnull() ]
```

```python
df['rating'] = df['rating'].astype(float)
```

### column : reviews

```python
df.head()
```

```python
df['reviews'].unique()
```

```python
df[ df['reviews'] == '3.0M' ]
```

```python
df.drop( index=df[ df['reviews'] == '3.0M' ].index[0] , inplace=True )
```

```python
df.shape
```

```python
df['reviews'] = df['reviews'].astype(int)
```

### column : size

```python
df['size'].unique()
```

### replace M and K with 0's

```python
# M = 1000000
# K = 1000
```

```python
def convert_to_millions(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return round(float(size.replace('k', '')) / 1000, 3)
    else:
        return 0  # or np.nan for missing/invalid entries like 'Varies with device'
```

```python
df['size_million'] = df['size'].apply(convert_to_millions)
```

```python
df.info()
```

```python
df.drop( columns='size' , inplace=True )
```

```python
df.head()
```

### column : price

```python
df['price'].unique()
```

```python
df['price_dollers'] = df['price'].str.replace('$','')
```

```python
df['price_dollers'] = df['price_dollers'].astype(float)
```

```python
df.info()
```

```python
df.drop(columns='price' , inplace=True)
```

```python
df.head()
```

### column : installs

```python
df['installs'].unique()
```

```python
df['installs'] = df['installs'].str.replace('+' , '')
df['installs'] = df['installs'].str.replace(',' , '')

df['installs'] = df['installs'].astype(float)
```

```python
df.info()
```

```python
df.head()
```

### column : current ver

```python
df['current_version'] = df['current ver'].replace( 'Varies with device' , '1.0.0' ) 
```

```python
df.drop(columns="current ver" , inplace=True)
```

```python
df.head()
```

### column : android ver

```python
df['android ver'].unique()
```

```python
df['android_version_required'] = df['android ver'].str.replace(' and up' , '')

df['android_version_required'] = df['android_version_required'].str.replace('W','')

df['android_version_required'] = df['android_version_required'].replace('Varies with device',0)

df['android_version_required'] = df['android_version_required'].replace(np.nan,0)
```

```python
df['android_version_required'] = df['android_version_required'].str.split('-').str[0]

df['android_version_required'] = df['android_version_required'].fillna(0)
```

```python
df['android_version_required'].unique()
```

```python
df.drop(columns='android ver' , inplace=True)
```

```python
df.head()
```

### column : last updated

```python
df['last_updated'] = pd.to_datetime(df['last updated'] )
```

```python
df.drop(columns='last updated' , inplace=True)
```

```python
df['year'] = df['last_updated'].dt.year

df['month'] = df['last_updated'].dt.month

df['day'] = df['last_updated'].dt.day
```

```python
df.info()
```

```python
df.head()
```

```python
numerical_features = [ feature for feature in df.columns if df[feature].dtype != 'object']

categorical_features = [ feature for feature in df.columns if df[feature].dtype == 'object']
```

```python
print(f'The numerical feature of the dataset are:{numerical_features}')

print(f'The numerical feature of the dataset are:{categorical_features}')
```

```python
for category in categorical_features:
    print(f'{df[category].value_counts(normalize=True)*100}')
    print('--------------------------------------------------------------------------------------------------------------------------------------')
```

```python
plt.figure(figsize=(15,15))
plt.suptitle( 'Univerient analysis of numerical columns' , fontsize=20 , fontweight='bold', alpha=0.8 , y = 1 )

for i in range(0 , len(numerical_features)):
    plt.subplot( 5,3,i+1 )
    sns.kdeplot(x = df[numerical_features[i]] , shade=True , color='r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

### Observation-2
### Reviwes , installs,size_million and price_doolers are in the left skewed
### ratings,last_updated,year are right skewed

```python
plt.figure(figsize=(10,10))
plt.suptitle( 'Univerient analysis of Categorical columns' , fontsize=20 , fontweight='bold', alpha=0.8 , y = 1 )

categories = ['type' , 'content rating']

for i in range(0,len(categories)):
    
    plt.subplot(2,2,i+1)
    
    sns.countplot( x= df[categories[i]] , palette='Set2' )
    
    plt.xlabel( categories[i].upper() )
    plt.xticks(rotation=45)
    plt.tight_layout()
```

### Observation-3
### In Type column : Free are more than paid
### In Content Rating : Everyone > Teen > Everyone 10+

```python

plt.figure(figsize=(15,5))

plt.title('Top 10 categories and their installation count')
sns.barplot(x=category , y=count , color='g')

plt.show()
```
